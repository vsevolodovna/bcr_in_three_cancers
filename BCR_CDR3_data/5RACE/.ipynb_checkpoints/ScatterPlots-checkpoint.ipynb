```python
import csv
import numpy as np
import pandas as pd
import os
import re
import shutil
import zipfile
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
from itertools import combinations

pat='lcp3'
izlist=['IGH','IGHA','IGHG','IGHM']
x_list=['LN12']
y_list=['LN12','LN13','tum11','tum12','tum13']

#top to normalize further
nhead=1000

#read data
patdata=pd.read_csv("C:\\Users\\User\\IBH\\Trees\\grouped_data\\"+pat+"_clone-pass.tsv",comment="#", sep="\t", header=0)
tum1_exp=pd.read_csv("C:\\Users\\User\\IBH\\Clonotype overlaps\\Heterogeneity_plots\\Tumor_Expanded\\clones_lcp3_tum1vsALL.csv",comment="#", sep=",", header=0)
tum2_exp=pd.read_csv("C:\\Users\\User\\IBH\\Clonotype overlaps\\Heterogeneity_plots\\Tumor_Expanded\\clones_lcp3_tum2vsALL.csv",comment="#", sep=",", header=0)
tum3_exp=pd.read_csv("C:\\Users\\User\\IBH\\Clonotype overlaps\\Heterogeneity_plots\\Tumor_Expanded\\clones_lcp3_tum3vsALL.csv",comment="#", sep=",", header=0)

tum1_exp_list=list(tum1_exp['rn'])
tum2_exp_list=list(tum2_exp['rn'])
tum3_exp_list=list(tum3_exp['rn'])

def isotype(s):
    if s.find('IGHA')!=-1:
        ans='IGHA'
    elif s.find('IGHG')!=-1:
        ans='IGHG'
    elif s.find('IGHM')!=-1:
        ans='IGHM'
    elif s.find('IGHD')!=-1:
        ans='IGHD'
    elif s.find('IGHE')!=-1:
        ans='IGHE'
    else: ans=s
    return ans
```

```python
patdata['patient']=[x.split('_')[1] for x in patdata['sequence_id']]
patdata=patdata[patdata['patient']==pat]
patdata['isotype']=[isotype(str(x)) for x in patdata['allchitswithscore']]
patdata['sample']=[x.split('_')[2]+x.split('_')[4]+x.split('_')[5] for x in patdata['sequence_id']]
patdata=patdata[patdata['sample'].isin(x_list) | patdata['sample'].isin(y_list)]
patdata['replicate']=[x.split('_')[6] for x in patdata['sequence_id']]
patdata['replicate_count']=patdata.groupby('sample')['replicate'].transform('nunique')
```

```python
for iz in izlist:
    df=patdata[patdata['isotype'].str.contains(iz)].copy()
    df=df[['aaseqcdr3','clonefraction','sample','replicate', 'replicate_count']]
    
     #suming up the same cdr3aa within one replica
    df['fraction_aa']=df.groupby(['aaseqcdr3','sample','replicate', 'replicate_count'])['clonefraction'].transform('sum')
    df=df.drop_duplicates(['aaseqcdr3','sample', 'replicate', 'replicate_count'])
    
    #Normalization of replicas size
    df['depth']=df.groupby(['sample','replicate'])['replicate'].transform('count')
    c=df['depth'].min()
    df=df.sort_values(['sample','replicate','fraction_aa'],ascending=False).groupby(['sample','replicate']).head(c)
    df['norm_coef']=df.groupby(['sample','replicate'])['fraction_aa'].transform('sum')
    df['fraction_aa']=df['fraction_aa']/df['norm_coef']
    
    #pooling replicas
    df['freq_pooled_repl']=df.groupby(['aaseqcdr3','sample'])['fraction_aa'].transform('sum')/df['replicate_count']
    df=df.drop_duplicates(['aaseqcdr3','sample'])
    
    #Cutting sample sizes to equal
    df['depth']=df.groupby(['sample'])['aaseqcdr3'].transform('count')
    c=min(df['depth'].min(),nhead)
    print(c)
    df=df.sort_values(['sample','freq_pooled_repl'],ascending=False).groupby(['sample']).head(c)
    df['norm_coef']=df.groupby(['sample'])['freq_pooled_repl'].transform('sum')
    df['freq_pooled_repl']=df['freq_pooled_repl']/df['norm_coef']
    
    df=df[['aaseqcdr3','sample','freq_pooled_repl']]
    
    for org1 in x_list:
        for org2 in y_list:
            if org1==org2:
                continue
            df_x=df[df['sample']==org1].copy()
            df_x.columns=['aaseqcdr3','sample','freq_x']
            df_y=df[df['sample']==org2].copy()
            df_y.columns=['aaseqcdr3','sample','freq_y']
            plot_data=pd.merge(df_x,df_y,how='outer',on='aaseqcdr3')
            plot_data=plot_data[['aaseqcdr3','freq_x','freq_y']]
            plot_data= plot_data.apply(lambda x: x.fillna(0.00005))
            
            #define color if expanded in part of the tumor
            plot_data['color_tum1']=["tum1 expanded" if x in tum1_exp_list else "not tum1 expanded" for x in plot_data['aaseqcdr3']]
            plot_data['color_tum2']=["tum2 expanded" if x in tum2_exp_list else "not tum2 expanded" for x in plot_data['aaseqcdr3']]
            plot_data['color_tum3']=["tum3 expanded" if x in tum3_exp_list else "not tum3 expanded" for x in plot_data['aaseqcdr3']]
            
            plot_data_color_tum1=plot_data[plot_data['color_tum1']=="tum1 expanded"]
            plot_data_color_tum2=plot_data[plot_data['color_tum2']=="tum2 expanded"]
            plot_data_color_tum3=plot_data[plot_data['color_tum3']=="tum3 expanded"]
            
            figure, axes = plt.subplots(nrows=1, ncols=3,figsize=(19, 5.5))
            axes[0].set_ylim(0.00004,0.5)
            axes[0].set_xlim(0.00004,0.5)
            axes[0].set_xscale('log')
            axes[0].set_yscale('log')
            sns.scatterplot(ax=axes[0],x="freq_x", y="freq_y",hue="color_tum1",alpha=0.3,data=plot_data,s=100, linewidth=0)
            sns.scatterplot(ax=axes[0],x="freq_x", y="freq_y",alpha=1,data=plot_data_color_tum1,s=100, linewidth=0,color='orange').set_title(pat+", "+iz+", top "+str(c))
            axes[0].set_xlabel(org1)
            axes[0].set_ylabel(org2)
            
            axes[1].set_ylim(0.00004,0.5)
            axes[1].set_xlim(0.00004,0.5)
            axes[1].set_xscale('log')
            axes[1].set_yscale('log')
            sns.scatterplot(ax=axes[1],x="freq_x", y="freq_y",hue="color_tum2",alpha=0.3,data=plot_data,s=100, linewidth=0)
            sns.scatterplot(ax=axes[1],x="freq_x", y="freq_y",alpha=1,data=plot_data_color_tum2,s=100, linewidth=0,color='orange').set_title(pat+", "+iz+", top "+str(c))
            axes[1].set_xlabel(org1)
            axes[1].set_ylabel(org2)
            
            axes[2].set_ylim(0.00004,0.5)
            axes[2].set_xlim(0.00004,0.5)
            axes[2].set_xscale('log')
            axes[2].set_yscale('log')
            sns.scatterplot(ax=axes[2],x="freq_x", y="freq_y",hue="color_tum3",alpha=0.3,data=plot_data,s=100, linewidth=0)
            sns.scatterplot(ax=axes[2],x="freq_x", y="freq_y",alpha=1,data=plot_data_color_tum3,s=100, linewidth=0,color='orange').set_title(pat+", "+iz+", top "+str(c))
            axes[2].set_xlabel(org1)
            axes[2].set_ylabel(org2)
```
